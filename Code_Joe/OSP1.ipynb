```python
def mostrar_menu():
    print("*********************************************\n\n")
    print("Bienvenido\n\n")
    print("1. Ingresar origen\n")
    print("2. Ingresar destino\n")
    print("3. Ingresar hora\n")
    print("4. Buscar tiempo de viaje medio\n")
    print("5. Salir\n\n")
    print("*********************************************\n\n")
```

```python
mostrar_menu()
```

```python
def menu():
  mostrar_menu()
  opcion = int(input())
  
  while(True):
      
    if(opcion == -1):
      mostrar_menu()
      opcion = int(input())

    if(opcion < -1 or opcion > 5):
      print("\nIngrese un valor valido\n")
      mostrar_menu()
      opcion = int(input())


    if(opcion == 1):
      print("\nIngrese ID del origen: \n")
      origen = int(input())
      while(1 > origen or origen > 1160):
        print("\nIngrese un valor valido\n")
        origen = int(input())
      opcion = -1
      

    if(opcion == 2):
      print("\nIngrese ID del destino: \n")
      destino = int(input())
      while(1 > destino or destino > 1160):
        print("\nIngrese un valor valido\n")
        destino = int(input())
      opcion = -1
    

    if(opcion == 3):
      print("\nIngrese hora del dÃ­a: \n")
      hora = int(input())
      while(0 > hora or hora > 23):
        print("\nIngrese un valor valido\n")
        hora = int(input())
      opcion = -1
    

    if(opcion == 4):
      print("Tiempo de viaje medio: ")

      idx = dt.loc[(dt['sourceid']==origen) & (dt['dstid']==destino) & (dt['hod'] == hora)].index.tolist()
      av_time = dt['mean_travel_time'].loc[idx]
      print(float(av_time))
      opcion = -1
 

    if(opcion == 5):
      return False
```
